```python
# Exercise 2: Docker Basics — Single Container

# Step 6: Pull and run Nginx
!docker run -d -p 8080:80 --name my-nginx nginx:alpine

# Screenshot

# Step 7: Inspect the container
inspect_output = !docker inspect my-nginx
print("Docker Inspect Output:")
print(inspect_output)

# Extract specific info (IP address, hostname, environment variables)
import json

inspect_output = !docker inspect my-nginx
inspect_json = json.loads("".join(inspect_output))

network_settings = inspect_json[0].get('NetworkSettings', {})
container_ip = network_settings.get('IPAddress', 'N/A')
container_hostname = inspect_json[0]['Config'].get('Hostname', 'N/A')
container_env = inspect_json[0]['Config'].get('Env', [])

print("Container IP Address:", container_ip)
print("Container Hostname:", container_hostname)
print("Container Environment Variables:", container_env)

# Step 8: Enter the container (using exec)
# You can run commands and capture outputs in Jupyter
hostname = !docker exec my-nginx hostname
print("\nHostname inside container:", hostname[0])

processes = !docker exec my-nginx ps aux
print("\nProcesses inside container:")
for line in processes:
    print(line)

os_info = !docker exec my-nginx cat /etc/os-release
print("\nOS Info inside container:")
for line in os_info:
    print(line)

# Step 9: Resource monitoring
print("\nDocker stats (run 5 seconds as example):")
!timeout 5 docker stats my-nginx --no-stream

# Step 10: Clean up
!docker stop my-nginx
!docker rm my-nginx
!docker rmi nginx:alpine

# Verify removal
ps_output = !docker ps -a
images_output = !docker images
print("\nDocker containers after cleanup:", ps_output)
print("\nDocker images after cleanup:", images_output)

# Step 11 (Challenge): Run two Nginx containers
!docker run -d -p 8080:80 --name nginx1 nginx:alpine
!docker run -d -p 8081:80 --name nginx2 nginx:alpine

print("\nTwo Nginx containers running:")
!docker ps

print("\nVerify accessibility: open http://localhost:8080 and http://localhost:8081 in browser")

# Clean up challenge containers
!docker stop nginx1 nginx2
!docker rm nginx1 nginx2
!docker rmi nginx:alpine
```
