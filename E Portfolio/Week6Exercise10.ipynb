```python
import socket
import threading

def handle_client(client_socket, addr):
    print(f"âœ… Connection from {addr}")
    data = client_socket.recv(1024)
    print(f"ðŸ“¨ Received: {data.decode()}")
    client_socket.sendall(data)  # Echo back
    client_socket.close()

def start_ipv4_server():
    s4 = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
    s4.bind(('localhost', 65432))
    s4.listen(5)
    print("ðŸŽ§ IPv4 Server listening on localhost:65432")
    
    while True:
        client_socket, addr = s4.accept()
        threading.Thread(target=handle_client, args=(client_socket, addr), daemon=True).start()

def start_ipv6_server():
    s6 = socket.socket(socket.AF_INET6, socket.SOCK_STREAM)
    s6.bind(('::1', 65433))
    s6.listen(5)
    print("ðŸŽ§ IPv6 Server listening on [::1]:65433")
    
    while True:
        client_socket, addr = s6.accept()
        threading.Thread(target=handle_client, args=(client_socket, addr), daemon=True).start()

# Start both servers in threads
threading.Thread(target=start_ipv4_server, daemon=True).start()
threading.Thread(target=start_ipv6_server, daemon=True).start()
```

```python
s = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
s.connect(('localhost', 65432))
s.sendall(b"Hello IPv4 Server!")
data = s.recv(1024)
print(f"Server echoed: {data.decode()}")
s.close()
```
