```python
import socket
import datetime

# Create TCP socket
server_socket = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
server_socket.bind(('localhost', 65432))
server_socket.listen(1)

print("üéß TCP Server listening on localhost:65432")

# Accept connection
client_socket, client_address = server_socket.accept()
print(f"‚úÖ Connection established with {client_address}")

# Start timing
start_time = datetime.datetime.now()

# Receive message
data = client_socket.recv(1024)
print(f"üì® Received: {data.decode()}")

# Echo back
client_socket.send(data)

# End timing
end_time = datetime.datetime.now()
time_taken = end_time - start_time
print(f"‚è± Time taken: {time_taken.total_seconds()} seconds")

# Close sockets
client_socket.close()
server_socket.close()
```

```python
import socket
import datetime

server_socket = socket.socket(socket.AF_INET, socket.SOCK_DGRAM)
server_socket.bind(('localhost', 65432))

print("üéß UDP Server listening on localhost:65432")

data, client_address = server_socket.recvfrom(1024)

start_time = datetime.datetime.now()

print(f"üì® Received: {data.decode()}")

server_socket.sendto(data, client_address)

end_time = datetime.datetime.now()
time_taken = end_time - start_time
print(f"‚è± Time taken: {time_taken.total_seconds()} seconds")

server_socket.close()
```
