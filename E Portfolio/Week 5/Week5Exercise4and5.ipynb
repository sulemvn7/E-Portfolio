```python
import socket
import threading
import datetime
import time
```

```python
def tcp_server():
    server_socket = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
    server_socket.bind(('localhost', 65432))
    server_socket.listen(1)
    print("üéß TCP Server listening on localhost:65432")

    client_socket, client_address = server_socket.accept()
    print(f"‚úÖ Connection established with {client_address}")

    start_time = datetime.datetime.now()

    with open('received_file_tcp.txt', 'wb') as f:
        while True:
            data = client_socket.recv(1024)
            if not data:
                break
            f.write(data)

    end_time = datetime.datetime.now()
    print(f"‚è± TCP file received in { (end_time - start_time).total_seconds() } seconds")

    client_socket.close()
    server_socket.close()
```

```python
def tcp_client():
    time.sleep(1)  # Give server a moment to start

    client_socket = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
    client_socket.connect(('localhost', 65432))

    start_time = datetime.datetime.now()

    with open('file_to_send.txt', 'rb') as f:
        client_socket.sendfile(f)

    end_time = datetime.datetime.now()
    print(f"‚è± TCP file sent in { (end_time - start_time).total_seconds() } seconds")

    client_socket.close()
```

```python
# Run server and client in parallel
threading.Thread(target=tcp_server, daemon=True).start()
threading.Thread(target=tcp_client, daemon=True).start()
```
