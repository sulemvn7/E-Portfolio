```python
from collections import deque

def round_robin(processes, time_quantum):
    """
    processes: list of (name, burst_time)
    time_quantum: max CPU slice per turn
    """

    queue = deque([(name, bt) for name, bt in processes])
    time_clock = 0
    gantt = []

    while queue:
        name, remaining = queue.popleft()

        run_time = min(remaining, time_quantum)

        gantt.append((name, time_clock, time_clock + run_time))

        time_clock += run_time
        remaining -= run_time

        if remaining > 0:
            queue.append((name, remaining))
        else:
            print(f"Process {name} completed at t={time_clock} ms")

    print("\nGantt Chart:")
    for name, start, end in gantt:
        print(f"| {name} ({start}-{end}ms) |", end="")
    print()


# Processes
procs = [('P1', 24), ('P2', 3), ('P3', 10)]

round_robin(procs, time_quantum=4)
```

```python
def fcfs(processes):

    time_clock = 0
    waiting_times = []
    gantt = []

    for name, burst in processes:

        start_time = time_clock
        end_time = time_clock + burst

        gantt.append((name, start_time, end_time))

        waiting_times.append(start_time)

        time_clock = end_time

        print(f"Process {name} completed at t={end_time} ms")

    avg_wait = sum(waiting_times) / len(waiting_times)

    print("\nGantt Chart:")
    for name, start, end in gantt:
        print(f"| {name} ({start}-{end}ms) |", end="")

    print("\n\nAverage Waiting Time:", avg_wait)


procs = [('P1', 24), ('P2', 3), ('P3', 10)]

fcfs(procs)
```

```python
def priority_scheduler(processes):

    # sort by priority (lower number = higher priority)
    processes.sort(key=lambda x: x[2])

    time_clock = 0

    print("Priority Scheduling Order:")

    for name, burst, priority in processes:

        start = time_clock
        end = time_clock + burst

        print(f"{name} (Priority {priority}) runs from {start} to {end}")

        time_clock = end


procs = [
    ('P1', 24, 2),
    ('P2', 3, 1),
    ('P3', 10, 3),
    ('P4', 15, 2)
]

priority_scheduler(procs)
```
