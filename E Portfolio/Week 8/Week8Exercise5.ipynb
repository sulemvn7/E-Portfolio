```python
import os
import stat
import time

def inspect_file(path):
    s = os.stat(path)

    print(f"Path : {path}")
    print(f"Size : {s.st_size} bytes")
    print(f"Inode : {s.st_ino}")
    print(f"Permissions: {oct(stat.S_IMODE(s.st_mode))}")
    print(f"Modified : {time.ctime(s.st_mtime)}")
    print(f"Is file : {stat.S_ISREG(s.st_mode)}")
    print(f"Is dir : {stat.S_ISDIR(s.st_mode)}")
    print()

for entry in os.scandir('.'):
    inspect_file(entry.path)
```

```python
latest_file = None
latest_time = 0

for entry in os.scandir('.'):
    mtime = os.stat(entry.path).st_mtime

    if mtime > latest_time:
        latest_time = mtime
        latest_file = entry.path

print("Most recently modified file:", latest_file)
print("Modified time:", time.ctime(latest_time))
```

```python
import time
import os

FILENAME = "test_io.bin"

DATA_SIZE = 10 * 1024 * 1024
chunk = b'X' * 4096

# Buffered write
t0 = time.perf_counter()

with open(FILENAME, "wb") as f:
    for _ in range(DATA_SIZE // len(chunk)):
        f.write(chunk)

t_buf = time.perf_counter() - t0

print("Buffered write:", round(t_buf,3), "seconds")

# Unbuffered write
t0 = time.perf_counter()

with open(FILENAME, "wb", buffering=0) as f:
    for _ in range(DATA_SIZE // len(chunk)):
        f.write(chunk)

t_unbuf = time.perf_counter() - t0

print("Unbuffered write:", round(t_unbuf,3), "seconds")

print("Speed-up:", round(t_unbuf / t_buf,2), "x faster")
```

```python
# Buffered read
t0 = time.perf_counter()

with open(FILENAME, "rb") as f:
    while f.read(4096):
        pass

read_buf = time.perf_counter() - t0

print("Buffered read:", round(read_buf,3), "seconds")

# Unbuffered read
t0 = time.perf_counter()

with open(FILENAME, "rb", buffering=0) as f:
    while f.read(4096):
        pass

read_unbuf = time.perf_counter() - t0

print("Unbuffered read:", round(read_unbuf,3), "seconds")

print("Read speed-up:", round(read_unbuf / read_buf,2), "x")
```

```python
import time

LOGFILE = "simple_log.txt"

def write_log(entry_id, message):

    with open(LOGFILE, "a") as f:
        timestamp = time.time()
        f.write(f"{entry_id},{timestamp},{message}\n")


def replay_log():

    print("Replaying log entries:\n")

    with open(LOGFILE, "r") as f:
        for line in f:
            entry_id, timestamp, message = line.strip().split(",")

            print("Entry:", entry_id)
            print("Time:", time.ctime(float(timestamp)))
            print("Message:", message)
            print()


write_log(1, "User login")
write_log(2, "File uploaded")
write_log(3, "System backup")

replay_log()
```
