```python
import os
import psutil

# Get current process info
current_pid = os.getpid()
process = psutil.Process(current_pid)

print("Process Name:", process.name())
print("PID:", current_pid)
print("Parent PID:", os.getppid())
print("Status:", process.status())
print("CPU %:", process.cpu_percent(interval=1))
print("Memory (MB):", process.memory_info().rss / 1024**2)
```

```python
print("\n--- Top 5 processes by CPU ---")

all_procs = [(p.pid, p.name(), p.cpu_percent(interval=0.1))
             for p in psutil.process_iter(['pid', 'name', 'cpu_percent'])]

top5 = sorted(all_procs, key=lambda x: x[2], reverse=True)[:5]

for pid, name, cpu in top5:
    print(f"PID {pid:6} {name:30} CPU: {cpu:.1f}%")
```

```python
python_processes = []

for p in psutil.process_iter(['pid', 'name']):
    if 'python' in p.info['name'].lower():
        python_processes.append((p.info['pid'], p.info['name']))

print("Python Processes Running:")

for pid, name in python_processes:
    print(f"PID {pid} - {name}")

print("\nTotal Python Processes:", len(python_processes))
```

```python
from multiprocessing import Process
import time
import os

def child_task():
    print("Child process running")
    print("Child PID:", os.getpid())
    time.sleep(5)

# Create child process
p = Process(target=child_task)

p.start()

print("Parent PID:", os.getpid())
print("Child started...")

p.join()

print("Child process finished")
```
