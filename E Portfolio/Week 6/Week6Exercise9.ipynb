```python
import ipaddress

def analyze_ip(ip_str):
    try:
        ip = ipaddress.ip_address(ip_str)
        print(f"IP: {ip.compressed}")
        print(f"Version: IPv{ip.version}")
        
        if ip.is_private:
            print("Type: Private")
        elif ip.is_loopback:
            print("Type: Loopback")
        elif ip.is_link_local:
            print("Type: Link-local")
        elif ip.is_multicast:
            print("Type: Multicast")
        else:
            print("Type: Public / Global Unicast")
    except ValueError:
        print("Invalid IP address!")

# Examples
analyze_ip("192.168.1.1")      # IPv4 private
analyze_ip("127.0.0.1")        # IPv4 loopback
analyze_ip("2001:db8::1")      # IPv6 global
analyze_ip("::1")               # IPv6 loopback
analyze_ip("fe80::1")           # IPv6 link-local
```
