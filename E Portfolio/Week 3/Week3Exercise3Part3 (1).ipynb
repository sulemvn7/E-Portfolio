```python
import subprocess

def tracert(domain):
    try:
        print(f"üó∫Ô∏è Tracking route to {domain}...")
        print("‚è≥ This takes 1-2 minutes - your data is globehopping!")
        print("=" * 60)

        result = subprocess.run(
            ["tracert", domain], 
            capture_output=True,
            text=True
        )

        print(result.stdout)

        print("=" * 60)
        print("‚úÖ Journey complete! Check out that route!")

    except FileNotFoundError:
        print("‚ùå tracert not found. Are you on Windows?")
        print("üí° Mac/Linux users: try 'traceroute' instead!")

    except Exception as e:
        print(f"‚ùå Something went wrong: {e}")


print("üó∫Ô∏è NETWORK PATHFINDER üó∫Ô∏è")
domain = input("Enter website or IP to track: ")
tracert(domain)
```
