```python
import socket
import time

def investigate_ip(website):
    try:
        start_time = time.time()  # Start timer
        
        ip_address = socket.gethostbyname(website)
        
        end_time = time.time()  # End timer
        lookup_time = end_time - start_time
        
        result = f"Website: {website}\nIP Address: {ip_address}\nLookup Time: {lookup_time:.6f} seconds\n"
        print(result)
        
        return result
    
    except socket.gaierror:
        error_message = f"Website: {website}\nError: Unable to resolve IP address\n"
        print(error_message)
        return error_message


# List of 3 websites
websites = ["google.com", "github.com", "bbc.co.uk"]

# Open file in write mode
with open("ip_investigation.txt", "w") as file:
    
    for site in websites:
        result = investigate_ip(site)
        file.write(result + "\n")

print("Investigation complete. Results saved to 'ip_investigation.txt'")
```
