```python
import socket
import requests

# Define locations with latitude and longitude
locations = {
    "University": {"lat": 51.7537, "lon": -1.2545},   # Example: University of Oxford
    "British Library": {"lat": 51.5299, "lon": -0.1273}  # British Library, London
}

weather_info = {}

for place, coords in locations.items():
    api_url = f"https://api.open-meteo.com/v1/forecast?latitude={coords['lat']}&longitude={coords['lon']}&current_weather=true"
    response = requests.get(api_url)
    
    if response.status_code == 200:
        weather_data = response.json()
        temperature = weather_data["current_weather"]["temperature"]
        weather_info[place] = temperature
        print(f"üå°Ô∏è {place}: {temperature}¬∞C")
    else:
        weather_info[place] = None
        print(f"‚ùå Failed to fetch weather data for {place}")
```

```python
import requests

locations = {
    "University": {"lat": 51.7537, "lon": -1.2545},
    "British Library": {"lat": 51.5299, "lon": -0.1273}
}

for place, coords in locations.items():
    api_url = f"https://api.open-meteo.com/v1/forecast?latitude={coords['lat']}&longitude={coords['lon']}&daily=temperature_2m_max,weathercode,windspeed_10m_max&timezone=Europe/London"
    response = requests.get(api_url)
    if response.status_code == 200:
        data = response.json()
        temp_max = data["daily"]["temperature_2m_max"][0]
        wind_max = data["daily"]["windspeed_10m_max"][0]
        weather_code = data["daily"]["weathercode"][0]
        print(f"{place}: Max Temp = {temp_max}¬∞C, Max Wind = {wind_max} km/h, Weather Code = {weather_code}")
    else:
        print(f"Failed to fetch weather for {place}")
```

```python
import socket

client_socket = socket.socket(socket.AF_INET, socket.SOCK_DGRAM)
server_address = ('localhost', 65433)

# Example message
message = f"University: Max Temp {temp_max}¬∞C, Max Wind {wind_max} km/h, Weather Code {weather_code}"

client_socket.sendto(message.encode(), server_address)
print("‚úÖ Weather data sent to server!")
client_socket.close()
```
