```python
import socket
import requests

# Define locations with latitude and longitude
locations = {
    "University": {"lat": 51.7537, "lon": -1.2545},   # Example: University of Oxford
    "British Library": {"lat": 51.5299, "lon": -0.1273}  # British Library, London
}

weather_info = {}

for place, coords in locations.items():
    api_url = f"https://api.open-meteo.com/v1/forecast?latitude={coords['lat']}&longitude={coords['lon']}&current_weather=true"
    response = requests.get(api_url)
    
    if response.status_code == 200:
        weather_data = response.json()
        temperature = weather_data["current_weather"]["temperature"]
        weather_info[place] = temperature
        print(f"ðŸŒ¡ï¸ {place}: {temperature}Â°C")
    else:
        weather_info[place] = None
        print(f"âŒ Failed to fetch weather data for {place}")
```

```python
if None not in weather_info.values():
    if weather_info["University"] > weather_info["British Library"]:
        print("ðŸŒž University is warmer today.")
    elif weather_info["University"] < weather_info["British Library"]:
        print("â„ï¸ British Library is warmer today.")
    else:
        print("ðŸŒ¡ï¸ Both locations have the same temperature.")
```

```python
# Prepare message
message = f"University: {weather_info['University']}Â°C, British Library: {weather_info['British Library']}Â°C"

# UDP Client
client_socket = socket.socket(socket.AF_INET, socket.SOCK_DGRAM)
server_address = ('localhost', 65433)  # Make sure your server is listening on this port
client_socket.sendto(message.encode(), server_address)
print("âœ… Weather data sent to server!")
client_socket.close()
```
